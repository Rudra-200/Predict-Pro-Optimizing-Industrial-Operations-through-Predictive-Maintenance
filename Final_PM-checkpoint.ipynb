By identifying patterns that precede equipment failures, the system should enable proactive maintenance, reducing unplanned downtime and maintenance expenses.

### Importing the necessary libraries for data preprocessing, visualization, model selection, and evaluation. ⚖️

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

import random
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import StackingClassifier

import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)
```

## Reading the CSV file into a DataFrame 📊📝👀

```python
data = pd.read_csv(r"predictive_maintenance.csv")
data.head(5)
```

## Dropping "Target" and "UDI" columns 🗑️💡

```python
data=data.drop(["Target"],axis = 1)
data=data.drop(["UDI"],axis = 1)
```

## 🔠 Encoding "Failure Type" column & 🗺️ Creating a class mapping dictionary

```python
le = LabelEncoder()

data["Failure Type"] = le.fit_transform(data["Failure Type"])

class_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

print("Class Mapping:", class_mapping)
```

## 🛠️ Defining features and target variable

```python
features = [feat for feat in data.columns if feat !='Failure Type']

X = data[features] # Feature set
y = data['Failure Type'] # Target
```

## 🚂🧪 Splitting the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
X_train.shape, X_test.shape
```

## 📊 Printing & Identifying numerical and categorical features

```python
print(f"The dataset contains {data.shape[0]} rows and {data.shape[1]} columns")

num_features = [feat for feat in features if data[feat].dtype != object]
cat_features = [feat for feat in features if data[feat].dtype == object]

print(f"Total number of features : {len(features)}")
print(f"Number of numerical features : {len(num_features)}")
print(f"Number of categorical features : {len(cat_features)}\n")
```

## Plotting boxplots for numerical features📊

```python
fig, axes = plt.subplots(5, 1, figsize=(8, 25))
for i, c in enumerate(num_features):
    f = data[[c]].boxplot(ax=axes[i], vert=False)
```

## 📊 Plotting Kernel Density Estimation (KDE) plots for numerical features

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 6))
df = data['Type'].value_counts()
barplot = df.plot(kind='pie', ax=axes[0], title='Type', autopct="%.2f", fontsize=14, ylabel='')
countplot = sns.countplot(x='Type', data=data, ax=axes[1])
```

## 📊 Calculating Pearson correlation matrix & Creating a heatmap

## 📊 Grouping by 'Type' and calculating the mean of 'Failure Type'

```python
df = data.groupby('Type')['Failure Type'].mean().to_frame().reset_index()
plot = df.plot(kind='bar', x='Type', y='Failure Type', color=('orange', 'skyblue'))
```

## MinMaxScaler object with a specified feature range of (0, 1), then fits and transforms the training set and transforms the test set using the scaler

```python
scaler = MinMaxScaler(feature_range=(0, 1))
X_train[num_features] = scaler.fit_transform(X_train[num_features]) #Fit and Transform the train set
X_test[num_features] = scaler.transform(X_test[num_features]) #Transform the test test
```

## Dropping 'Product ID' and 'Type' columns 🫳

```python
X_train.drop(['Product ID', 'Type'], axis=1, inplace=True)
X_test.drop(['Product ID', 'Type'], axis=1, inplace=True)
```

## 🌳 Creating a DecisionTreeClassifier object🏛️

```python
tree = DecisionTreeClassifier(random_state=1)
tree.fit(X_train, y_train)

print("Train accuracy : ", accuracy_score(y_train, tree.predict(X_train)))
print("Test accuracy : ", accuracy_score(y_test, tree.predict(X_test)))
```

```python
# 🎯 Define hyperparameters
distribution = {'max_depth': [4, 6, 8, 10, 12, 14, 16],
                'criterion': ['gini', 'entropy'],
                'min_samples_split': [2, 10, 20, 30, 40],
                'max_features': [0.2, 0.4, 0.6, 0.8, 1],
                'max_leaf_nodes': [8, 16, 32, 64, 128, 256],
                'class_weight': [{0: 1, 1: 2}, {0: 1, 1: 3}, {0: 1, 1: 4}, {0: 1, 1: 5}]
               }

# 🎯 Random search for best hyperparameters
search = RandomizedSearchCV(DecisionTreeClassifier(random_state=1),
                         distribution,
                         scoring='accuracy',
                         cv=3,
                         verbose=1,
                         random_state=1,
                         n_iter=30)

# 🎯 Fit the randomized search
search.fit(X_train, y_train)

# 🎯 Best parameters for Decision Tree classifier
search.best_params_
```

```python
# Retrain with best model

best_tree = search.best_estimator_

best_tree.fit(X_train, y_train)
print(" Best train accuracy : ", accuracy_score(y_train, best_tree.predict(X_train)))
print(" Best test accuracy : ", accuracy_score(y_test, best_tree.predict(X_test)))
```

```python
print(classification_report(y_test, best_tree.predict(X_test), zero_division=0.0))
```

## 📈 Creating a K-Nearest Neighbors (KNN) model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# 📊 Printing train and test accuracies
print("Train accuracy:", accuracy_score(y_train, knn.predict(X_train)))
print("Test accuracy:", accuracy_score(y_test, knn.predict(X_test)))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Hyperparameters for KNN
param_grid = {'n_neighbors': range(1, 21), 
              'weights': ['uniform', 'distance'], 
              'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'], 
              'leaf_size': [10, 20, 30, 40, 50], 
              'p': [1, 2]}

# Random search for best hyperparameters
search = RandomizedSearchCV(KNeighborsClassifier(),
                            param_grid,
                            scoring='accuracy',
                            cv=3,
                            verbose=1,
                            random_state=1,
                            n_iter=30)

# Fit the randomized search
search.fit(X_train, y_train)

# Print best parameters for KNN classifier
search.best_params_
```

```python
# Retrain with best model
best_knn = search.best_estimator_

# 📈 Fitting the best KNN model on the training data
best_knn.fit(X_train, y_train)

# 📊 Printing best train and test accuracies
print("Best train accuracy:", accuracy_score(y_train, best_knn.predict(X_train)))
print("Best test accuracy:", accuracy_score(y_test, best_knn.predict(X_test)))
```

```python
print(classification_report(y_test, best_knn.predict(X_test), zero_division=0.0))
```

## 📈 Creating a Support Vector Classifier (SVC) model

```python
svc = SVC(random_state=1)

svc.fit(X_train, y_train)

print("Train accuracy : ", accuracy_score(y_train, svc.predict(X_train)))
print("Test accuracy : ", accuracy_score(y_test, svc.predict(X_test)))
```

```python
# Hyperparameters
param_grid = {'C': [0.1, 1, 10, 100, 1000], 
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['linear', 'rbf', 'poly'],
              'degree':[0, 1, 2, 3, 4, 5, 6]
             } 

# Random search for best hyperparameters
search = RandomizedSearchCV(SVC(random_state=1),
                         param_grid,
                         scoring='accuracy',
                         cv=3,
                         verbose=1,
                         random_state=1,
                         n_iter=30)

search.fit(X_train, y_train)

# Best parameters for Support vector classifier
search.best_params_
```

```python
# Retrain with best model

best_svc = search.best_estimator_

best_svc.fit(X_train, y_train)
print("Best train accuracy : ", accuracy_score(y_train, best_svc.predict(X_train)))
print("Best test accuracy : ", accuracy_score(y_test, best_svc.predict(X_test)))
```

```python
print(classification_report(y_test, best_svc.predict(X_test), zero_division=0.0))
```

## 🌳 Creating a RandomForestClassifier model

```python
# Hyperparameters
params_grid = {'bootstrap': [True, False],
             'max_depth': [2, 5, 10, 20, None],
             'max_features': ['auto', 'sqrt'],
             'min_samples_leaf': [1, 2, 4],
             'min_samples_split': [2, 5, 10],
             'n_estimators': [50, 100, 150, 200]}

# Random search for best hyperparameters
search = RandomizedSearchCV(RandomForestClassifier(random_state=1),
                         params_grid,
                         scoring='accuracy',
                         cv=3,
                         verbose=1,
                         random_state=1,
                         n_iter=20)

search.fit(X_train, y_train)

# Best parameters for Random forest classifier
search.best_params_
```

```python
# Retrain with best model

best_forest = search.best_estimator_

best_forest.fit(X_train, y_train)
print("Best train accuracy : ", accuracy_score(y_train, best_forest.predict(X_train)))
print("Best test accuracy : ", accuracy_score(y_test, best_forest.predict(X_test)))
```

```python
print(classification_report(y_test, best_forest.predict(X_test),zero_division=0.0))
```

## 🧱 Creating a Stacking Classifier

```python
stack = StackingClassifier(estimators=[('best tree classifier', best_tree),
                                       ('best knn', best_knn),
                                       ('best svc', best_svc),
                                       ('best forest classifier', best_forest)],
                           
                           final_estimator=best_forest,
                           passthrough=True)

stack.fit(X_train, y_train)

print("Train accuracy : ", accuracy_score(y_train, stack.predict(X_train)))
print("Test accuracy : ", accuracy_score(y_test, stack.predict(X_test)))
```

```python
print(classification_report(y_test, stack.predict(X_test),zero_division=0.0))
```

```python
from sklearn.metrics import confusion_matrix

# Assuming 'stack' is your trained StackingClassifier
y_pred_prob = stack.predict_proba(X_test)

# Get the predicted class labels
y_pred = stack.predict(X_test)

# Create confusion matrices for each class
class_names = ['Heat Dissipation Failure', 'No Failure','Overstrain Failure', 'Power Failure', 'Random Failures','Tool Wear Failure']
for i, class_name in enumerate(class_names):
    # Filter the predictions and ground truth labels for the current class
    y_true_class = (y_test == i)  # Assuming class labels start from 0
    y_pred_class = (y_pred == i)
    
    # Calculate confusion matrix for the current class
    cm = confusion_matrix(y_true_class, y_pred_class)
    
    # Plot the confusion matrix (optional)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted False', 'Predicted True'],
                yticklabels=['Actual False', 'Actual True'])
    plt.title(f'Confusion Matrix for {class_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('Actual Label')
    plt.show()

    # Print the confusion matrix
    print(f'Confusion Matrix for {class_name}:')
    print(cm)
```

```python
def predict_failure_type(air_temp, process_temp, speed, torque, tool_wear):
    input_data = scaler.transform([[air_temp, process_temp, speed, torque, tool_wear]])
    prediction = stack.predict(input_data)[0]
    return le.inverse_transform([prediction])[0]

# Example of using the function to make predictions
air_temp = float(input("Enter Air temperature [K]: "))
process_temp = float(input("Enter Process temperature [K]: "))
speed = float(input("Enter Rotational speed [rpm]: "))
torque = float(input("Enter Torque [Nm]: "))
tool_wear = float(input("Enter Tool wear [min]: "))

predicted_failure_type = predict_failure_type(air_temp, process_temp, speed, torque, tool_wear)
print("Predicted Failure Type:", predicted_failure_type)
```
